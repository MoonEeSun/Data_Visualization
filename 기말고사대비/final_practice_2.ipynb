```python
new_iris = iris[['sepal_length', 'species']].groupby('species').apply(lambda x: x.mean()) # sepal_length 
new_iris.sort_values('sepal_length', inplace=True) #   sort
new_iris.reset_index(inplace=True)

# Draw plot
fig, ax = plt.subplots(figsize=(16,10), dpi= 80)

# 
ax.vlines(x=new_iris.index, ymin=0, ymax=new_iris.sepal_length, color='firebrick', alpha=0.7, linewidth=2)

#   :: s  - marker size
ax.scatter(x=new_iris.index, y=new_iris.sepal_length, s=75, color='firebrick', alpha=0.7)

# Title, Label, Ticks and Ylim
ax.set_title('sepal_length Lollipop Chart', fontdict={'size':22})
ax.set_ylabel('sepal_length ')
ax.set_xticks(new_iris.index)
ax.set_xticklabels(new_iris.species.str.lower(), rotation=30, fontdict={'horizontalalignment': 'right', 'size':12})

#      
for row in new_iris.itertuples():
    ax.text(row.Index, row.sepal_length+.3, s=round(row.sepal_length, 2), horizontalalignment= 'center', verticalalignment='bottom', fontsize=14)
```

```python
new_titanic = titanic[['fare', 'embark_town']].groupby('embark_town').apply(lambda x: x.mean())
new_titanic.sort_values('fare', inplace=True) # 
new_titanic.reset_index(inplace=True)

print('new_titanic')
print(new_titanic)

print('new_titanic.index')
print(new_titanic.index)

# Draw plot
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(16,10), facecolor='white', dpi= 80)
ax.vlines(x=new_titanic.index, ymin=0, ymax=new_titanic.fare, color='firebrick', alpha=0.7, linewidth=20) # linewidth  barchart

# Annotate Text : Bar     .
for i, income in enumerate(new_titanic.fare):
    ax.text(i, income+0.5, round(income, 1), horizontalalignment='center') #      round(, 1)

#  , y , c Ticks 
ax.set_title('Ordered Bar Chart', fontdict={'size':22})
ax.set(ylabel='embark_town  ', ylim=(0, 66))
plt.xticks(new_titanic.index, new_titanic.embark_town.str.lower(), rotation=60, horizontalalignment='right', fontsize=12)
```

```python
import squarify

new_titanic = titanic.groupby('class').size().reset_index(name='counts') #  :  
print('new_titanic')
print(new_titanic)

labels = new_titanic.apply(lambda x: str(x[0]) + "\n (" + str(x[1]) + ")", axis=1) # \n    
print('labels') #     
print(labels)

sizes = new_titanic['counts'].values.tolist()

print('sizes')
print(sizes)

colors = [plt.cm.Spectral(i/float(len(labels))) for i in range(len(labels))]

print('colors')
print(colors)

# Draw Plot
plt.figure(figsize=(12,8), dpi= 80)
squarify.plot(sizes=sizes, label=labels, color=colors, alpha=.8)

# Decorate
plt.title('titanic  class  Treemap')
plt.axis('off')
```
