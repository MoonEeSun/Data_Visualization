### seaborn
- barplot : x축 범주형 데이터, y축 수치형 데이터
- countplot : 범주형 데이터 (y축 데이터 X)
- histplot : 일변량 수치형 데이터 (y축 데이터 X)
- distplot : 일변량(1D) 데이터 (y축 데이터 X)
    - kde, rug, hist 한 번에 보여줌
- scatterplot : x, y축 수치형 데이터 
- jointplot : x, y축 수치형 데이터
    - scatter, hex, reg, kde 중 하나 가운데에, 히스토그램 가장자리에
- heatmap : 2차원 카테고리 데이터 - pivot_table 사용 (x, y축 데이터 X)
- pairplot : 3차원 이상의 실수형 데이터 (x, y축 데이터 X)
- kdeplot : x, y축 수치형 데이터 (y축 없어도 됨)
- lmplot : x, y축 수치형 데이터
- replot : x, y축 수치형 데이터
- boxplot : x축 범주형 데이터, y축 수치형 데이터
- violinplot : x축 범주형 데이터, y축 수치형 데이터 (y축 없어도 됨)
- stripplot : x축 범주형 데이터, y축 수치형 데이터
- swarmplot : x축 범주형 데이터, y축 수치형 데이터
- catplot : sns.catplot(data=df, kind='option', col='강조할 열', col_wrap = '개수')

```python
# import
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# darkgrid 효과 넣고 싶으면 넣기
sns.set(style = 'darkgrid') # style : white, dark, whitegrid, darkgrid,ticks
plt.rc('font', family = 'AppleGothic') # Mac용 - 한글 폰트 깨짐을 방지

# data load
iris = sns.load_dataset('iris')

sns.lmplot(x='sepal_length', y='petal_length', data=iris, hue = 'species')
plt.xlim(4.1 , 8.1)
plt.show()
```
