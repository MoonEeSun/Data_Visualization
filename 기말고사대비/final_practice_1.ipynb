```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

# 한글깨짐 방지
plt.rc('font', family='AppleGothic')
```

```python
titanic = sns.load_dataset('titanic')
iris = sns.load_dataset('iris')
setosa = iris.query("species == 'setosa'")
virginica = iris.query("species == 'virginica'")
versicolor = iris.query("species == 'versicolor'")
```

```python
# barplot(1)
sns.barplot(x='who', y='age', hue = 'survived', data =titanic, errwidth = 0)
plt.title('성별에 따른 생존율 그래프')
plt.xlabel('성별')
plt.ylabel('나이')
```

```python
# barplot(2)
sns.barplot(y='who', x='fare', hue='survived', data=titanic)
```

```python
# countplot(1)
sns.countplot(x='species', data=iris)
plt.title('iris 종의 개수')
```

```python
sns.countplot(x='class', data=titanic, hue='survived')
plt.title('클래스별 생존율')
```

```python
# histplot(1) 
sns.histplot(x=setosa['sepal_length'], kde = True, bins= 5)
plt.title('iris의 setosa의 sepal length')
```

```python
# histplot(2)
sns.histplot(x='age', data=titanic, kde=True)
plt.title('타이타닉 탑승객 나이 분포')
```

```python
# distplot(1)
sns.distplot(x=titanic['age'], rug=True, kde =True, hist=True, color='r')
plt.title('titanic호 탑승객 나이 분포')
```

```python
# distplot(2)
sns.distplot(x=setosa['sepal_length'], vertical=True, hist= True, rug= True, kde=True)
plt.title('iris의 setosa sepal length 분포')
```

```python
# scatter(1)
sns.scatterplot(x='age', y='fare', data=titanic)
plt.xlabel('나이')
plt.ylabel('운임요금')
plt.title('타이타닉 나이에 따른 운임요금 분포')
```

```python
# scatter(2)
sns.scatterplot(x='age', y='fare', data=titanic)
sns.rugplot(x='age', y='fare', data=titanic, color='r')
plt.xlabel('나이')
plt.ylabel('운임요금')
plt.title('타이타닉 나이에 따른 운임요금 분포')
```

```python
# joint plot(1)
sns.jointplot(x='fare', y='age', data=titanic, color='b', kind='hex')
```

```python
# heat map(1)
pivot = titanic.pivot_table(index='class', columns='who', values='survived')
sns.heatmap(data=pivot, annot=True)
plt.title('성별에 따른 클래스와 생존여부')
```

```python
# pair plot(1)
titanic_3d=titanic.loc[:, ['survived', 'fare', 'age', 'pclass']]
sns.pairplot(data=titanic_3d, hue='survived')
```

```python
# kdeplot(1)
sns.kdeplot(x='fare', y='age', data=titanic, shade=True)
plt.title('운임요금에 따른 나이')
```

```python
# kdeplot(2)
sns.kdeplot(x='sepal_length', data=setosa)
sns.rugplot(x='sepal_length', data=setosa)
plt.title('iris의 setosa sepal length에 대한 rug plot / kde 곡선')
plt.ylabel('값')
```

```python
# kdeplot(3)
sns.kdeplot(x='sepal_length',data=setosa)
sns.kdeplot(x='sepal_length',data=virginica)
sns.kdeplot(x='sepal_length',data=versicolor)
plt.legend(['setosa', 'virginica', 'versicolor'])
```

```python
# lmplot(1)
sns.lmplot(x='age', y='fare', hue='sex', data=titanic)
```

```python
# lmplot(2)
sns.lmplot(x='age', y='fare', hue='sex', data=titanic)
plt.xlim(0,70)
plt.ylim(0,250)
```

```python
# relplot(1)
sns.relplot(x='age', y='fare', hue='sex', row='who', col='survived', data=titanic)
```

```python
# boxplot(1)
new_titanic = titanic.loc[titanic['fare']<=100]
sns.boxplot(x='class', y='fare', data=new_titanic)
```

```python
# boxplot(2)
sns.boxplot(x='class', y='age', hue='survived', data=titanic)
plt.ylim(10,50)
```

```python
# boxplot(3)
sns.boxplot(x='class', y='fare', data=new_titanic, whis=np.inf)
sns.stripplot(x='class', y='fare', data=new_titanic, color='0.4', jitter=True)
```

```python
# violinplot(1)
sns.violinplot(x='sepal_length', data=iris)
```

```python
# violinplot(2)
sns.violinplot(data=iris)
plt.title('iris')
```

```python
# violinplot(3)
sns.violinplot(x='species', y= 'sepal_length', data=iris)
```

```python
# violinplot(4)
sns.violinplot(x='class', y='age', data=titanic, inner=None)
sns.swarmplot(x='class', y='age', data=titanic, color='0.9')
```

```python
# violinplot(5)
sns.boxplot(data=iris)
sns.violinplot(data=iris)
```

```python
# stripplot(1)
sns.stripplot(x='class', y='fare', data=titanic, jitter=True, size= 3)
plt.ylim(0,100)
```

```python
# stripplot(2)
sns.stripplot(x='class', y='fare', hue='sex', data=titanic, jitter=True, size=3, dodge=True)
plt.ylim(10,100)
```

```python
# swarmplot(1)
sns.swarmplot(x='class', y='age', data=titanic)
plt.ylim(10, 70)
```

```python
#swarmplot(2)
sns.swarmplot(x='class', y='age', data=titanic, hue='survived')
```

```python
# catplot(1)
sns.catplot(x='class', y='survived', hue='sex', kind='bar', data=titanic)
```

```python
# catplot(2)
g = sns.catplot(data=titanic, col='survived', kind='bar')
g.fig.set_size_inches(20, 10)
g.fig.subplots_adjust(wspace=0.3, hspace=0.3)
```

```python
new_titanic = titanic.loc[:, ['survived','age','fare', 'class']]
sns.catplot(data=new_titanic, col='class', kind='box', col_wrap=2)
plt.ylim(0,100)
```

### matplotlib

```python
fig = plt.figure(figsize=(8, 5))
data1 = titanic.loc[:, ['class', 'fare']]
plt.plot(data1['class'], data1['fare'],'-.sm', markersize=10, markerfacecolor=(0.8, 0.5, 0.8))
data2 = titanic.loc[:, ['class', 'age']]
plt.bar(data2['class'], data2['age'], color='y')
plt.title('클래스별 요금과 나이 현황')
plt.ylim(0,100)
plt.xlabel('클래스')
plt.legend(['요금', '나이'])
```

```python
data = titanic.loc[:,['age', 'fare', 'class', 'who', 'survived']]
sns.catplot(data=data, col_wrap=3, kind='box', col='who')
plt.ylim(0,80)
```
