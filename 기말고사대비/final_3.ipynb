```python
# import
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# darkgrid 효과 넣고 싶으면 넣기
sns.set(style = 'darkgrid') # style : white, dark, whitegrid, darkgrid,ticks
plt.rc('font', family = 'AppleGothic') # Mac용 - 한글 폰트 깨짐을 방지

# data load
iris = sns.load_dataset('iris')

sns.distplot(x=iris['sepal_width'], rug=True, hist=True, kde=True, color = 'red')
plt.xlim(1.4, 4.99)
plt.ylim(0, 1.4)
plt.ylabel('')
plt.xlabel('sepal_width')
plt.show()
```
