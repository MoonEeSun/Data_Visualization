```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family = 'AppleGothic')
```

```python
titanic = sns.load_dataset('titanic')
iris = sns.load_dataset('iris')
print(titanic.info())
print(iris.info())
```

```python
new_titanic = titanic[['age', 'embark_town']].groupby('embark_town').apply(lambda x: x.mean()) # sepal_length의 평균
new_titanic.sort_values('age', inplace=True) # 평균을 오름차순으로 sort
new_titanic.reset_index(inplace=True)

# Draw plot
fig, ax = plt.subplots(figsize=(16,10), dpi= 80)

# 수직선
ax.vlines(x=new_titanic.index, ymin=0, ymax=new_titanic.age, color='firebrick', alpha=0.7, linewidth=2)

# 상단의 점 :: s 파라미터 - marker size
ax.scatter(x=new_titanic.index, y=new_titanic.age, s=75, color='firebrick', alpha=0.7)

# Title, Label, Ticks and Ylim
ax.set_title('embark town에서의 age의 Lollipop Chart', fontdict={'size':22})
ax.set_ylabel('나이의 평균')
ax.set_xticks(new_titanic.index)
ax.set_xticklabels(new_titanic.embark_town.str.lower(), rotation=30, fontdict={'horizontalalignment': 'right', 'size':12})

# 상단에 그려 주는 소득 정보 텍스트
for row in new_titanic.itertuples():
    ax.text(row.Index, row.age+.3, s=round(row.age, 2), horizontalalignment= 'center', verticalalignment='bottom', fontsize=14)
```

```python
new_iris = iris[['sepal_length', 'species']].groupby('species').apply(lambda x: x.mean()) # sepal_length의 평균
new_iris.sort_values('sepal_length', inplace=True) # 평균을 오름차순으로 sort
new_iris.reset_index(inplace=True)

# Draw plot
fig, ax = plt.subplots(figsize=(16,10), dpi= 80)

# 수직선
ax.vlines(x=new_iris.index, ymin=0, ymax=new_iris.sepal_length, color='firebrick', alpha=0.7, linewidth=20)

# 상단의 점 :: s 파라미터 - marker size
# ax.scatter(x=new_iris.index, y=new_iris.sepal_length, s=75, color='firebrick', alpha=0.7)

# Title, Label, Ticks and Ylim
ax.set_title('sepal_length의 Lollipop Chart', fontdict={'size':22})
ax.set_ylabel('sepal_length의 평균')
ax.set_xticks(new_iris.index)
ax.set_xticklabels(new_iris.species.str.lower(), rotation=30, fontdict={'horizontalalignment': 'right', 'size':12})

# 상단에 그려 주는 소득 정보 텍스트
for row in new_iris.itertuples():
    ax.text(row.Index, row.sepal_length+.1, s=round(row.sepal_length, 2), horizontalalignment= 'center', verticalalignment='bottom', fontsize=14)
```
