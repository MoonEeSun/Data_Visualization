```python
# import
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.rc('font', family = 'AppleGothic') # Mac용 - 한글 폰트 깨짐을 방지

# data load
titanic = sns.load_dataset('titanic')

sns.stripplot(x='class', y='age', data=titanic, hue='survived', dodge=True, size = 2, jitter=True)
plt.show()
```
