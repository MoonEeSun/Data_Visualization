```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 폰트
plt.rc('font', family = 'AppleGothic')

# 그래프를 이미지로 저장할 때
cnt, PNG, UNDERBAR = 0, '.png', '_'
CHART_NAME = 'Broken_plot'
```

```python
def FileSave():
    global cnt
    cnt +=1
    savefile = CHART_NAME + UNDERBAR + str(cnt).zfill(2) + PNG
    plt.savefig(savefile, dpi = 400)
    print(savefile + '파일이 저장되었습니다.')
```

# 단일 꺾은선 그래프

```python
# data load
data1 = pd.read_csv('../data/주요발생국가주간동향(4월2째주).csv', index_col = '국가')
print(data1)
chart_data1 = data1['4월06일']
print(chart_data1)


# draw chart
fig1 = plt.figure()
YTICKS_INTERVAL = 50000 # Y축 눈금에 그려질 간격
maxlim = (int(chart_data1.max()) / YTICKS_INTERVAL + 1) * YTICKS_INTERVAL # 최대 눈금의 값 설정 - 올림
plt.plot(chart_data1, color = 'blue', linestyle = 'solid', marker = 'o') # broken line chart

values = np.arange(0,  maxlim + 1, YTICKS_INTERVAL) # ytick
plt.yticks(values)
plt.grid(True) # grid
plt.xlabel('국가명') # X label
plt.ylabel('발생 건수') # Y label
plt.title('4월 6일 코로나 발생 건수') # 제목
# plt.show()
FileSave()
```

# 이중축 꺾은선 그래프

```python
# data load
data3 = pd.read_csv('../data/tips.csv')
print(data3)
# data - 결제 금액과 팁만 사용
data_bill = data3.loc[range(100), ['total_bill']]
data_tip = data3.loc[range(100), ['tip']]


# draw plot
## axis 1
fig4, ax1 = plt.subplots() # .subplots()의 형이 튜플이므로 fig는 저장하기 위해

ax1.set_title('결제 금액과 Tip(이중축)')
color = 'tab:red'
ax1.set_ylabel('결제 금액', color = color)
ax1.plot(range(0,100), data_bill, color = color, label = 'total_bill') # 내용 그리기! plot(x_data, y_data)
ax1.tick_params(axis = 'y', labelcolor = color) # y축 글씨 빨강색으로


## axis 2
ax2 = ax1.twinx() # ax1과 같은 축이므로 .twinx() 사용

color = 'tab:blue'
ax2.set_ylabel('팁(tip)', color=color)
ax2.plot(range(0, 100), data_tip, color=color, label = 'tip')
ax2.tick_params(axis = 'y', labelcolor=color)


fig4.tight_layout()
# plt.show()
FileSave()
```

```python
from matplotlib.backends.backend_pdf import PdfPages

pdfsave = PdfPages(CHART_NAME +"_19011773.pdf")
pdfsave.savefig(fig1)
pdfsave.savefig(fig2) 
pdfsave.savefig(fig3)## 안됨
pdfsave.savefig(fig4)

pdfsave.close()
```
