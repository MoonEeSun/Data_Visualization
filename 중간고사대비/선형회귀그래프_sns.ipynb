```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
plt.rc('font', family='AppleGothic')
plt.rcParams["font.size"] = 12.
plt.rcParams['xtick.labelsize'] = 12.
plt.rcParams['ytick.labelsize'] = 12.
plt.rcParams['axes.unicode_minus'] = False
```

```python
CHART_NAME = 'Lmplot_sns'
cnt, PNG, UNDERBAR = 0, '.png', '_'
filename = 'welfareClean.csv'
```

```python
# 파일 저장

def FileSave():
    global cnt
    cnt +=1
    savefile = CHART_NAME + UNDERBAR + str(cnt).zfill(2) + PNG
    plt.savefig(savefile, dpi = 400)
    print(savefile + '파일이 저장되었습니다.')
```

# 기본 lmplot

```python
# data load
welfare = pd.read_csv(filename, encoding = 'CP949')

## 1) 기본 lmplot
fig1 = plt.figure()
fig1 = sns.lmplot(x = '나이', y='소득', height = 8, data =welfare)
plt.title('기본 lmplot')

# fig1.fig.subplots_adjust(top=.9)
# fig1.fig.suptitle('기본 lmplot')
# y축의 값이 300인 부근에 존재하는 직선은 회귀선을 의미

# plt.show()
FileSave()
```

# 범주형 데이터의 항목별로 선형관계 그리기

```python
# data load
welfare = pd.read_csv(filename, encoding = 'CP949')

## 2) 범주형 데이터의 항목별로 선형관계 그리기
fig2 = plt.figure()
fig2 = sns.lmplot(x = '나이', y='소득', hue = '결혼 유무', height = 8, data = welfare)
plt.title('')


# 이혼에 대한 데이터가 상대적으로 적음

# plt.show()
FileSave()
```

# col 매개변수를 사용하여 항목별 그래프를 별도로 그려보기

```python
# data load
welfare = pd.read_csv(filename, encoding = 'CP949')

## 3) col 매개변수를 사용하여 항목별 그래프를 별도로 그려보기
# col_wrap으로 한 줄에 표기할 column 개수 명시
fig3 = plt.figure()
fig3 = sns.lmplot(x = '나이', y='소득', hue = '결혼 유무', col = '성별', col_wrap = 2, height = 8, data = welfare)
# plt.title('성별로 구분하는 lmplot')
fig3.fig.subplots_adjust(top=.9)
fig3.fig.suptitle('성별로 구분하는 lmplot')

# plt.show()
FileSave()
```

```python
from matplotlib.backends.backend_pdf import PdfPages 

pdfsave = PdfPages(CHART_NAME + "_19011773.pdf")
pdfsave.savefig(fig1.fig)
pdfsave.savefig(fig2.fig)
pdfsave.savefig(fig3.fig)
pdfsave.close()
```
