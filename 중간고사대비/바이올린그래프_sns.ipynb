```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
plt.rc('font', family='AppleGothic')
plt.rcParams["font.size"] = 12.
plt.rcParams['xtick.labelsize'] = 12.
plt.rcParams['ytick.labelsize'] = 12.
plt.rcParams['axes.unicode_minus'] = False
```

```python
CHART_NAME = 'Violin_plot_sns'
cnt, PNG, UNDERBAR = 0, '.png', '_'
filename = 'welfareClean.csv'
```

```python
# 파일 저장

def FileSave():
    global cnt
    cnt +=1
    savefile = CHART_NAME + UNDERBAR + str(cnt).zfill(2) + PNG
    plt.savefig(savefile, dpi = 400)
    print(savefile + '파일이 저장되었습니다.')
```

# 기본

```python
# data load
welfare = pd.read_csv(filename, encoding = 'CP949')

fig1= plt.figure()
plt.title('나이에 따른 바이올린 그래프')
sns.violinplot(x = '나이', data = welfare)

# plt.show()
FileSave()

# 약 50정도의 나이에 많은 데이터가 분포되어 있음을 알 수 있음
```

# x, y축 지정 - 바이올린 분할

```python
# data load
welfare = pd.read_csv(filename, encoding = 'CP949')

fig2 = plt.figure()
plt.title('비교 분포 확인')
sns.violinplot(x = '성별', y = '나이', data = welfare)

# plt.show()
FileSave()
```

# 가로형태의 바이올린 그래프

```python
# data load
welfare = pd.read_csv(filename, encoding = 'CP949')

fig3 = plt.figure()
plt.title('가로형 violinplot')
sns.violinplot(y = '성별', x = '나이', data = welfare)


# plt.show()
FileSave()
```

# Hue 옵션으로 분포 비교

```python
# data load
welfare = pd.read_csv(filename, encoding = 'CP949')

fig4 = plt.figure()
plt.title('hue 옵션으로 분포 비교')
sns.violinplot(x = '종교 유무', y= '나이', hue = '성별', data = welfare, palette = "muted")

# plt.show()
FileSave()
```

```python
from matplotlib.backends.backend_pdf import PdfPages 

pdfsave = PdfPages(CHART_NAME + "_19011773.pdf")
pdfsave.savefig(fig1)
pdfsave.savefig(fig2)
pdfsave.savefig(fig3)
pdfsave.savefig(fig4)
pdfsave.close()
```
