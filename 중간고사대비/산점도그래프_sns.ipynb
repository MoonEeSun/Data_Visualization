```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
plt.rc('font', family='AppleGothic')
plt.rcParams["font.size"] = 12.
plt.rcParams['xtick.labelsize'] = 12.
plt.rcParams['ytick.labelsize'] = 12.
plt.rcParams['axes.unicode_minus'] = False
```

```python
CHART_NAME = 'Scatter_plot_sns'
cnt, PNG, UNDERBAR = 0, '.png', '_'
filename = 'welfareClean.csv'
```

```python
# 파일 저장

def FileSave():
    global cnt
    cnt +=1
    savefile = CHART_NAME + UNDERBAR + str(cnt).zfill(2) + PNG
    plt.savefig(savefile, dpi = 400)
    print(savefile + '파일이 저장되었습니다.')
```

# 기본 replot

```python
# data load
welfare = pd.read_csv(filename, encoding = 'CP949')


## 1) 기본 replot
fig1 = plt.figure()
fig1 = sns.relplot(x='나이', y= '소득', hue='성별', data=welfare)
plt.title('기본 repllot')


# plt.show()
FileSave()
```

# 결혼 유무 범주형 데이터 개수 만큼 데이터 늘리기 - col 옵션

```python
# data load
welfare = pd.read_csv(filename, encoding = 'CP949')


## 2) 결혼 유무 범주형 데이터 개수 만큼 데이터를 늘리기 - col 옵션
fig2 = plt.figure()
fig2 = sns.relplot(x='나이', y='소득', hue='성별', col='결혼 유무', data =welfare)
plt.title('col 옵션으로 그래프 분할')


# plt.show()
FileSave()
```

# 행과 열 둘다 늘리기 - row 옵션도 사용

```python
# data load
welfare = pd.read_csv(filename, encoding = 'CP949')


## 3) row 옵션도 사용
# row -> y축  
# col -> x축
fig3 = plt.figure()
fig3 = sns.relplot(x='나이', y='소득', hue='성별', col='결혼 유무', row='연령대', data =welfare)
plt.title('행과 열이 늘어난 산점도 그래프')


# plt.show()
FileSave()
```

# 색상 지정

```python
# data load
welfare = pd.read_csv(filename, encoding = 'CP949')


## 4) 색상 지정
fig4 = plt.figure()
fig4 = sns.relplot(x='나이', y='소득', hue='성별', col='결혼 유무', palette='CMRmap_r', data =welfare)
plt.title('컬러 팔레트 적용')
# 색 팔레트 지정 :: palette

# plt.show()
FileSave()
```

```python
from matplotlib.backends.backend_pdf import PdfPages 

pdfsave = PdfPages(CHART_NAME + "_19011773.pdf")
pdfsave.savefig(fig1.fig)
pdfsave.savefig(fig2.fig)
pdfsave.savefig(fig3.fig)
pdfsave.savefig(fig4.fig)
pdfsave.close()
```
