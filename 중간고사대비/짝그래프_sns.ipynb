```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
plt.rc('font', family='AppleGothic')
plt.rcParams["font.size"] = 12.
plt.rcParams['xtick.labelsize'] = 12.
plt.rcParams['ytick.labelsize'] = 12.
plt.rcParams['axes.unicode_minus'] = False
```

```python
CHART_NAME = 'Pairplot_sns'
cnt, PNG, UNDERBAR = 0, '.png', '_'
filename = 'welfareClean.csv'
```

```python
# 파일 저장

def FileSave():
    global cnt
    cnt +=1
    savefile = CHART_NAME + UNDERBAR + str(cnt).zfill(2) + PNG
    plt.savefig(savefile, dpi = 400)
    print(savefile + '파일이 저장되었습니다.')
```

# 짝그래프 - 3차원 이상의 실수형 데이터

# 기본 pairplot

```python
# data load
welfare = pd.read_csv(filename, encoding = 'CP949')
newwelfare = welfare.loc[:, ['직업 코드', '소득', '나이', '결혼 유무']]

fig1 = plt.figure()
fig1 = sns.pairplot(data = newwelfare)
# plt.title('기본 pairplot')
fig1.fig.subplots_adjust(top=.9)
fig1.fig.suptitle('기본 pairplot')

# plt.show()
FileSave()

# 우측 하단의 '나이'에 대한 히스토그램을 확인해보면 대략 50대에서 60대가 많이 분포되어 있음을 확인
# 중앙의 '소득'에 대한 히스토그램을 보면 유독 큰 소득이 있는데, 이상치일 확률이 높음
```

# 결혼 유무(범주형데이터)에 따른 그래프 추출
- 범주형 데이터는 색상으로 구분됨

```python
# data load
welfare = pd.read_csv(filename, encoding = 'CP949')
newwelfare = welfare.loc[:, ['직업 코드', '소득', '나이', '결혼 유무']]

fig2 = plt.figure()
fig2 = sns.pairplot(data=newwelfare, hue= '결혼 유무')
# plt.title('hue 옵션으로 특성 구분')
fig2.fig.subplots_adjust(top=.9)
fig2.fig.suptitle('hue 옵션으로 특성 구분')

# plt.show()
FileSave()
```

# 다른 색상을 지정

```python
# data load
welfare = pd.read_csv(filename, encoding = 'CP949')
newwelfare = welfare.loc[:, ['직업 코드', '소득', '나이', '결혼 유무']]

fig3 = plt.figure()
fig3= sns.pairplot(data=newwelfare, hue = '결혼 유무', palette = 'rainbow')
# plt.title('컬러 팔레트 적용')
fig3.fig.subplots_adjust(top=.9)
fig3.fig.suptitle('컬러 팔레트 적용')


# plt.show()
FileSave()
```

# height 높이 지정

```python
# data load
welfare = pd.read_csv(filename, encoding = 'CP949')
newwelfare = welfare.loc[:, ['직업 코드', '소득', '나이', '결혼 유무']]

fig4 = plt.figure()
fig4 = sns.pairplot(data=newwelfare, hue = '결혼 유무', palette = 'rainbow', height = 5)
# plt.title('사이즈 적용')
fig4.fig.subplots_adjust(top=.9)
fig4.fig.suptitle('사이즈 적용')

# plt.show()
FileSave()
```

```python
from matplotlib.backends.backend_pdf import PdfPages

pdfsave = PdfPages(CHART_NAME + "_19011773.pdf")
pdfsave.savefig(fig1.fig)
pdfsave.savefig(fig2.fig)
pdfsave.savefig(fig3.fig)
pdfsave.savefig(fig4.fig)
pdfsave.close()
```
