## 1. Matplotlib 라이브러리

- 산점도 그래프

```python
# x축과 y축 좌표 설정
x = data['이름']
y = data['연령']

# 산점도 그래프 환경 설정 (전)
plt.scatter(x, y)

# 산점도 그래프 환경 설정 (후)
plt.figure(figsize=(8, 5))
plt.scatter(x, y)
plt.title('혈압측정 참가자의 연령 분포 현황', fontsize=15)
plt.xlabel('Name')
plt.ylabel('Age')
plt.ylim(0, 80)
plt.legend(['연령'])
# plt.savefig('1_혈압측정 참가자의 연령 분포 현황.png')
```

### 그래프 수정 옵션 :: 선스타일-마커-색상
- 선 스타일
    - '-' : 실선
    - '--' : 파선
    - ':' : 점선
    - '-.' : 일점 쇄선
- 마커 기호 지정
    - '+' : 플러스 기호
    - 'o' : 원
    - '*' : 별
    - '.' : 점
    - '+' : 십자형
    - 'x', 'X' : 엑스자형
    - 's', 'square' : 정사각형
    - 'd', 'diamond' : 다이아몬드형
    - '^' : 위쪽 방향 삼각형
    - '<' : 왼쪽 방향 삼각형
    - '>' : 오른쪽 방향 삼각형
    - 'v' : 아래쪽 방향 삼각형
    - 'p', 'pentagram' : 오각형
    - 'h', 'hexagram' : 육각형
- 색상 지정
    - r : 빨
    - g : 녹
    - b : 파
    - c : 녹청
    - m : 자홍
    - y : 노란
    - k : 검정
    - w : 흰


```python
plt.figure(figsize=(8, 5))

x1 = data['이름']
y1 = data['연령']
plt.plot(x1, y1, '*m', markersize = 15, markerfacecolor=[0.1, 0.6, 0.6]) # '--*m' -> '*m'

x2 = data['이름']
y2 = data['BMI']
plt.plot(x2, y2, '-.or', markersize = 13, markeredgecolor = 'r', markerfacecolor = 'y')

plt.title('연령별 BMI 수치 현황', fontsize = 15)
plt.xlabel('참가자 이름')
plt.ylabel('연령 분포')
plt.ylim(0,100)
plt.legend(['연령', 'BMI'])
```

```python
plt.figure(figsize=(8,5))

x1 = data['이름']
y1 = data['키']
plt.plot(x1, y1, '-.sm', markersize = 10, markerfacecolor=[0.8, 0.5, 0.8])

x2 = data['이름']
y2 = data['체중']
plt.bar(x2, y2, color = 'y')

plt.title('연령별 키와 체중 수치 현황')
plt.legend(['키', '체중'])
```

## 4. titanic & iris dataset

### titanic

```python
df = sns.load_dataset('titanic')
print(df.shape)
print(df.head())
print(df.columns)
```

```python
# 클래스별로 구분하여 인원 정보 출력
df.groupby('class').count()
```

```python
# 클래스별 탑승인원의 생존율 분석
print(df.groupby('class')['survived'].count())

print(df.pivot_table('survived', index = 'class', columns = 'who'))
```

- 클래스별 생존율 바 플롯 시각화

```python
# 그래프 스타일 설정
sns.set(style = 'darkgrid')
plt.rc('font', family='AppleGothic')

sns.catplot(data=df, x='class', y = 'survived', hue = 'sex', kind='bar')
plt.title('타이타닉 탑승 인원의 생존율 분석 현황', fontsize=13)
plt.show();
```

- displot : hist + kdeplot
    - 일변량 데이터 (수치형 데이터)

```python
x = df['age']
sns.distplot(x, rug=True, hist=True, kde=True)
```

- heatmap
    - 데이터가 2차원 & 모든 값이 카테고리(범주형)

```python
pivot = df.pivot_table(index = 'class', columns='who', values='survived')
print(pivot)
sns.heatmap(data=pivot, annot=True)
plt.show()
```

- pairplot
    - 3차원 이상의 다차원 실수형 데이터
    - Dataset 중 2개 컬럼간 관계

```python
new_df = df.loc[:, ['survived', 'fare', 'age', 'pclass']]
print(new_df)
sns.pairplot(data=new_df, hue='survived')
plt.show()
```

- lmplot

```python
sns.lmplot(x='age', y='fare', hue='sex', height=8, data=df)
```

- relplot

```python
sns.relplot(x='age', y='fare', hue='sex', col='survived', row='who', data=df)
```

- jointplot
    - 중앙에 산점도, 가장자리에 히스토그램
    - 수치형 데이터
    - kind 옵션
        - 'scatter' : 산점도
        - 'hex' : 육각형 모양의 밀도
        - 'reg': 회귀선
        - 'kde' : 밀집도

```python
sns.jointplot(x='fare', y='age', data=df, kind='hex')
```

```python
sns.barplot(x='who', y = 'age', hue='survived', data=df, errwidth=0)
```

```python
# 성별에 따른 생존율?
sns.barplot(x='who', y='survived', hue='sex', data=df)
```

```python
sns.boxplot(x='sex', y='age', data=df)
```

### iris

```python
df = sns.load_dataset('iris')
print(df.head())
print(df.shape)
print(df.columns)
print(df['species'].unique())
print(df.describe())
```

```python
# 아이리스 데이터 프레임 부분 데이터 셋
setosa = df.query("species == 'setosa'")
virginica = df.query("species == 'virginica'")
versicolor = df.query("species == 'versicolor'")
print(setosa.describe())

plt.rc('font', family='AppleGothic')
```

- 히스토그램과 커널밀도시각화

```python
sns.distplot(setosa['sepal_length'], hist = True, kde = True, rug = True)
```

- 복수개의 커널밀도

```python
sns.kdeplot(setosa['sepal_length'])
sns.kdeplot(virginica['sepal_length'])
sns.kdeplot(versicolor['sepal_length'])
```

```python
# plt.rc.set_text(s, 0.0, flags=flags)
# sns.kdeplot(data = setosa['sepal_length'], data2 = virginica['sepal_length'])
```

- countplot: 범주형 데이터에 대하여 항목별 개수를 세서 막대 그래프 그림

```python
sns.countplot(x='species', data = df)
```

- violinplot : KDE + Box

```python
sns.violinplot(x='sepal_length', data=df)
```

```python
sns.violinplot(x='species', y = 'sepal_length', data=df)
```

- stripplot : 일변량 데치터셋에 대해서 간략히 요약
    - x, y : x, y data
    - data : dataset
    - hue : 특성 구분 컬럼
    - dodge : hue 매개변수 사용시, dodge = True면 범주형 데이터 축을 따라서 strips 데이터 분리
    - jitter : True면 데이터 개수 많을 경우 겹치지 않도록 해서 데이터 분포 이해 도움
    - size : 마커의 크기 지정

```python
sns.stripplot(x='species', y = 'sepal_length', data=df, jitter=True)
```

- swarmplot : stripplot()과 유사
    - 점들이 보정됨 : 서로 겹쳐지지 않음

```python
sns.swarmplot(x='species', y = 'sepal_length', data=df, size=2)
```
