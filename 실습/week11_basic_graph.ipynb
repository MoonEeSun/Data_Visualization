# pandas를 이용하여 그래프 그리기
## kind 옵션
- 선 그래프 : df.plot(kind='line'), df.plot()
- 수직 막대 그래프 : df.plot(kind='bar')
- 수평 막대 그래프 : df.plot(kind='barh')
- 히스토그램 : df.plot(kind='hist')
- 박스 플롯 : df.plot(kind='bax')
- 커널 밀도 그래프 : df.plot(kind='kde')
- 면적 그래프 : df.plot(kind='area')
- 파이 그래프 : df.plot(kind='pie')
- 산점도 그래프 : df.plot(kind='scatter')
- 고밀도 산점도 그래프 : df.plot(kind='hexbin')

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rc('font', family='AppleGothic') # 한글 폰트 깨짐 해결
data = '../data/systolic.csv'
df = pd.read_csv(data)
```

### 1. 선 그래프
- 시간의 경과에 따라 데이터에 의한 변화의 상황 나타냄

```python
df.plot()
```

```python
new_df = df.iloc[[0,1,2], 8:]
new_df.index=['신영웅', '송가인', '오대양']

new_df.plot(kind='line')
```

### 2. 막대 그래프
- 각 항목의 수량과 크기를 비교하기 쉽고 시각적으로 데이터셋 전체를 쉽게 파악
- 데이터 값의 크기를 정확히 나타낼 수 있다

```python
new_df = df.iloc[[0,1,2], 8:]
new_df.index=['신영웅', '송가인', '오대양']

new_df.plot(kind='bar')
```

```python
new_df = df.iloc[[0,1,2], 5:8]
new_df.index=['신영웅', '송가인', '오대양']

new_df.plot(kind='barh')
```

### 3. 박스 플롯
- 데이터의 특성을 시각화

```python
new_df = df.iloc[[0,1,2], 8:]
new_df.index=['신영웅', '송가인', '오대양']

new_df.plot(kind='box')
```

### 커널 밀도 그래프
- 수집된 정보를 **많은** 데이터들의 분포로부터 원래 변수의 확률 분포의 특성을 추정하고자 할 때

```python
new_df = df.iloc[[0,1,2], 8:]
new_df.index=['신영웅', '송가인', '오대양']

new_df.plot(kind='kde')
```

```python
new_df = df.iloc[:5, 8:]
new_df.index=['이겨례', '강산애','신영웅', '송가인', '오대양']

new_df.plot(kind='box')
new_df.plot(kind='kde')
```

### 그 외의 그래프

```python
new_df = df.iloc[[0,1,2], 8:]
new_df.index=['신영웅', '송가인', '오대양']

new_df.plot(kind = 'hist', x = '1차') # histogram
new_df.plot(kind = 'area') # 면적 그래프
new_df.plot(kind = 'pie', y = '4차') # pie graph


new_df = df.iloc[[0,1,2], 8:]
new_df.index=['신영웅', '송가인', '오대양']
new_df.plot(kind = 'scatter', x = '1차', y = '3차') # 산점도 그래프
new_df.plot(kind = 'hexbin', x = '3차', y = '1차') # 고밀도 산점도 그래프
```
